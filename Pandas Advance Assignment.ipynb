```python
Q1. Here are five functions of the pandas library with execution:

python
Copy code
import pandas as pd

# Function 1: Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22]}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# Function 2: Describing the DataFrame
description = df.describe()
print("\nDataFrame Description:")
print(description)

# Function 3: Selecting a column
ages = df['Age']
print("\nSelected Column 'Age':")
print(ages)

# Function 4: Sorting by a column
sorted_df = df.sort_values(by='Name')
print("\nSorted DataFrame by 'Name':")
print(sorted_df)

# Function 5: Filtering rows based on a condition
filtered_df = df[df['Age'] > 25]
print("\nFiltered DataFrame where 'Age' > 25:")
print(filtered_df)
Q2. Function to re-index the DataFrame:

python
Copy code
def reindex_dataframe(df):
    new_index = range(1, len(df) * 2, 2)
    df_reindexed = df.set_index(pd.Index(new_index))
    return df_reindexed

# Example usage:
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
reindexed_df = reindex_dataframe(df)
print("Re-indexed DataFrame:")
print(reindexed_df)
Q3. Function to calculate the sum of the first three values in the 'Values' column:

python
Copy code
def sum_of_first_three(df):
    values_sum = df['Values'].head(3).sum()
    print("Sum of the first three values:", values_sum)

# Example usage:
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
sum_of_first_three(df)
The rest of the questions follow a similar pattern. I'll be happy to help you with any specific function or question!




User
please give all answer 
ChatGPT
Certainly! Here are the answers to the remaining questions:

Q4. Function to create a new column 'Word_Count':

python
Copy code
def word_count(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df

# Example usage:
df = pd.DataFrame({'Text': ['This is a sample text', 'Another example', 'Just a few words']})
df = word_count(df)
print("DataFrame with Word_Count column:")
print(df)
Q5. DataFrame.size() vs DataFrame.shape():

DataFrame.size(): Returns the number of elements in the DataFrame (total number of cells).
DataFrame.shape(): Returns a tuple representing the dimensions of the DataFrame (number of rows, number of columns).
Q6. Function to read an Excel file:

python
Copy code
def read_excel_file(file_path):
    df = pd.read_excel(file_path)
    return df

# Example usage:
excel_file_path = 'path/to/your/file.xlsx'
df = read_excel_file(excel_file_path)
print("DataFrame from Excel file:")
print(df)
Q7. Function to extract 'Username' from 'Email' column:

python
Copy code
def extract_username(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df

# Example usage:
df = pd.DataFrame({'Email': ['john.doe@example.com', 'alice.smith@example.com']})
df = extract_username(df)
print("DataFrame with Username column:")
print(df)
Q8. Function to select rows based on conditions:

python
Copy code
def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Example usage:
df = pd.DataFrame({'A': [3, 8, 6, 2, 9], 'B': [5, 2, 9, 3, 1], 'C': [1, 7, 4, 5, 2]})
selected_df = select_rows(df)
print("Selected DataFrame:")
print(selected_df)
Q9. Function to calculate mean, median, and standard deviation:

python
Copy code
def calculate_stats(df):
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_deviation = df['Values'].std()
    print("Mean:", mean_value)
    print("Median:", median_value)
    print("Standard Deviation:", std_deviation)

# Example usage:
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
calculate_stats(df)
Q10. Function to create a 'MovingAverage' column:

python
Copy code
def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df

# Example usage:
df = pd.DataFrame({'Sales': [10, 15, 20, 25, 30, 35, 40, 45, 50]})
df = calculate_moving_average(df)
print("DataFrame with MovingAverage column:")
print(df)
Q11. Function to create a 'Weekday' column:

python
Copy code
def add_weekday_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df

# Example usage:
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})
df = add_weekday_column(df)
print("DataFrame with Weekday column:")
print(df)
Q12. Function to select rows based on date range:

python
Copy code
def select_rows_by_date_range(df, start_date, end_date):
    df['Date'] = pd.to_datetime(df['Date'])
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    return selected_rows

# Example usage:
df = pd.DataFrame({'Date': ['2023-01-15', '2023-01-20', '2023-01-25', '2023-02-01']})
selected_df = select_rows_by_date_range(df, '2023-01-01', '2023-01-31')
print("Selected DataFrame:")
print(selected_df)
Q13. To use the basic functions of pandas, NumPy needs to be imported first.

python
Copy code
import pandas as pd
import numpy as np
```
